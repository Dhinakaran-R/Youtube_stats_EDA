# **Data Cleaning**

# **Removing Unwanted Columns**

 1. From above, I observed that  "Youtber" and "Title" shares the same info.we can keep either one of them, likewise "category" and "channel_type" shares the same info, we can eliminate one of them

2. Abbreviation,created_month,created_date,Gross tertiary education enrollment (%),Unemployment rate,Urban_population,Latitude,Latitude - these features can be removed as they arent helpful for analysis.

```python
#removing the above mentioned columns
data = data.drop(['Title','category','Abbreviation','created_month','created_date',\
                  'Gross tertiary education enrollment (%)','Unemployment rate','Urban_population','Latitude','Longitude'],axis = 1)
data.head(2)
```

# **Transforming and renaming columns**

1. Since columns like
  'subscribers',
  'video views',
  'video_views_for_the_last_30_days',
  'highest_monthly_earnings',
  'lowest_yearly_earnings',
  'highest_yearly_earnings',
  'lowest_monthly_earnings',
  'Population'
are represented in bigger numbers, lets transform them to millions in count for better understanding in analysis



```python
# Creating a function to transform and rename the above mentione columns

def convert_million(x,df):
  for i in x:
    df[i]=round(df[i]/1000000,2)
    df.rename(columns={i:f'{i}_in_ millions'},inplace=True)
  return df

#passing values to the function
cols = ['subscribers','video views','video_views_for_the_last_30_days','highest_monthly_earnings',\
       'lowest_yearly_earnings','highest_yearly_earnings','lowest_monthly_earnings','subscribers_for_last_30_days','Population']


data = convert_million(cols,data)

# viewing the transformed and renamed columns
data.head(1)
```

# **Points to ponder :**

1. Numerical columns such as country_rank,channel_type_rank,video_views for last 30 days,subscriber for last 30 days imputing them with mean or median is inappropriate.
2. since there can be channels which are newly started or not having subscribers since they dont have good contents,or may not be having video views for last 30 days since they wouldnt have uploaded any videos in last 30 days.
3. in case of real time projects we should reach out to clients to get the whole data for the column having null values.since this is out of scope in our practice, so its better to impute all the null values with **'0'** for Quantitative columns

4. categorical columns which are having null values can be replaced  by 'unknown' as i go with the above logic.




```python
# filling null values of numerical columns with 0 and categorical columns with 'unknown'
for i in data.select_dtypes(include = ['int64','float64']).columns:
  data[i] = data[i].fillna(0)

for i in data.select_dtypes(include= 'object').columns:
  data[i] = data[i].fillna('unknown')

```

```python
#verifying the missing values after imputation

data.isna().sum().to_frame(name = 'missing value count')
```

# **Data types Review**

```python
# As mentioned earlier converting data type of columns for ease of use.

#function to convert float columns to int
def float_to_int(x,df):
  for i in x:
    df[x] = df[x].astype('int64')
  return df

#passing values to the function
x = ['video_views_rank','country_rank','channel_type_rank','video_views_for_the_last_30_days_in_ millions','created_year']
data = float_to_int(x,data)

#coverting the data type of created_year column to object dtype
data['created_year'] = data['created_year'].astype(str)

#verifying data types after conversion
data.dtypes
```

# **Removing / Treating Wrong values**

```python
# rows having spl chars in Youtuber col such as- ï¿½ý imputed as 'Unknown'


import re

data ['Youtuber'] = data['Youtuber'].apply(lambda x :re.sub(r'[¿½ï|ý]+','',x))
def imputer(x):
  return x.strip()

data['Youtuber'] = data['Youtuber'].apply(imputer)

a = data ['Youtuber'].value_counts().index[0]
U_index = data[data['Youtuber'] == a].index
data.loc[U_index,'Youtuber'] = 'Unknown'
```

```python
# removing rows which are having 0 views though having subscribers and uploads
index = data[(data['video views_in_ millions']==0) & (data['uploads']!=0)].index
data = data.drop(index=index,axis = 0)

index1 = data[(data['video views_in_ millions']==0 & (data['subscribers_in_ millions']!=0))].index
data = data.drop(index=index1,axis = 0)
```
